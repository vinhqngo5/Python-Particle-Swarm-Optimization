# Giới thiệu thuật toán Particle swarm optimization

Trong khoa học máy tính, thuật toán Particle swarm optimization là một phương pháp tối ưu bằng cách lặp đi lặp lại cải thiện kết quả của những ứng viên thông qua những thước đo chất lượng nhất định

Thuật toán giải quyết một vấn đề bằng cách khởi tạo một quần thể các ứng cử viên, được gọi là Particle (hạt) và di chuyển các hạt này xung quanh không gian tìm kiếm theo các công thức toán học đơn giản về vị trí và vận tốc của hạt

Chuyển động của mỗi hạt sẽ bị ảnh hưởng bởi vị trí hiện tại của nó, vị trí tốt nhất nó từng đạt được, vị trí tốt nhất của những hạt khác -> Kì vọng sẽ đưa quần thể đến những giải pháp tốt nhất
